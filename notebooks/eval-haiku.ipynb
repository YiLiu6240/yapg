```python
ROOT = find_project_root()
```

```python
MODEL_NAME = settings.english_bert_model_name
CHECKPOINT_PATH = ROOT / settings.path_to_haiku_checkpoint
assert CHECKPOINT_PATH.exists()
```

```python
flags = ""
parser = create_parser()
args = parser.parse_args(flags)
print(args)
```

```python
tokenizer = BertTokenizerFast.from_pretrained(MODEL_NAME)
tokenizer
```

```python
finetuned_model = ModelModule(args=args).load_from_checkpoint(
    str(CHECKPOINT_PATH), args=args
)
finetuned_model
```

```python
starting_text = "winter"
num_poems = 20

poems = []
for i in range(num_poems):
    text = generate_haiku_text(
        tokenizer=tokenizer,
        model=finetuned_model,
        starting_text=starting_text,
        max_syllable_length=12,
    )
    print(text)
    poems.append(poems)
```
